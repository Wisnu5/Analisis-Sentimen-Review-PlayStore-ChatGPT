```python
from google_play_scraper import Sort, reviews_all
import pandas as pd
app_id = 'com.openai.chatgpt'

result = reviews_all(
    app_id,
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.NEWEST
)

dt = pd.DataFrame(result)
dt_gpt = dt[['content', 'score']]
```
